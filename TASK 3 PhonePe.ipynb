TASK 3: DATA QUALITY CHECKS
    --> Ensure data consistency across state and district levels.
    --> Display the discripancies found between state-level and district-level data.

```python
import pandas as pd

# STATE WISE AGGREGATE DATA FROM DISTRICT DATASET
d_txnusers=pd.read_excel(r"C:\Users\sonali\Downloads\PhonePayData.xlsx",sheet_name='District_Txn and Users')
result1=d_txnusers.groupby('State').agg({
    'Transactions':'sum',
    'Amount (INR)':'sum',
    'Registered Users':'sum'}).round(0).reset_index()

# STATE WISE AGGREGATE DATA FROM STATE DATASET
s_txnusers=pd.read_excel(r"C:\Users\sonali\Downloads\PhonePayData.xlsx",sheet_name='State_Txn and Users')
result2=s_txnusers.groupby('State').agg({
    'Transactions':'sum',
    'Amount (INR)':'sum',
    'Registered Users':'sum'}).round(0).reset_index()

# MERGE BOTH TABLES 
result=result1.merge(result2,on='State',suffixes=('_district','_state'))

# NOW WE WILL COMPUTE DISCREPANCIES
result['Transactions_diff']=result['Transactions_district']-result['Transactions_state']
result['Amount (INR)_diff']=result['Amount (INR)_district']-result['Amount (INR)_state']
result['Registered Users_diff']=result['Registered Users_district']-result['Registered Users_state']

# DISPLAY ONLY WITH DISCREPANCIES
discrepancies=result[(result['Transactions_diff']!=0) | (result['Amount (INR)_diff']!=0) | (result['Registered Users_diff']!=0)][
    ['State','Transactions_diff','Amount (INR)_diff','Registered Users_diff']]
print(discrepancies)
```

       --> It looks like there are some missing data in in state data.
