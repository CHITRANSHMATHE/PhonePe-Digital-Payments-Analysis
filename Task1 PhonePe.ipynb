```python
import pandas as pd
s_txnusers=pd.read_excel(r"C:\Users\sonali\Downloads\PhonePayData.xlsx",sheet_name='State_Txn and Users')

# Check total of null values in each column
print(s_txnusers.isna().sum())

# Calculating % of missing values in each column 
missing_percent=(s_txnusers.isna().mean()/len(s_txnusers))*100
missing_percent=missing_percent[missing_percent>0]
print("The % missing values in a columns are: ")
print(missing_percent)
if not missing_percent.empty:
    print("The column with highest % of missing values is:")
    print(missing_percent.idxmax())
else:
    print("No missing values found in the dataset.")
```

1.4 Calculate the total number of States and Districts. Identify state with highest number of districts.

```python
import pandas as pd
d_demo=pd.read_excel(r"C:\Users\sonali\Downloads\PhonePayData.xlsx",sheet_name='District Demographics')

#Total number of States and Districts
total_states=d_demo['State'].nunique()
total_districts=d_demo['District'].nunique()
print(f"Total number of States are: {total_states}")
print(f"Total number of Districts are: {total_districts}")

# State with highest number of Districts
dist_per_state=d_demo.groupby('State')['District'].nunique()
hdist=dist_per_state.idxmax()
htotal=dist_per_state.max()
print(f"The state of {hdist} has the highest number of districts {htotal}.")
```
